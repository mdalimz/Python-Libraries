```python
import pandas as pd
```

```python
diamond=pd.read_csv('D:/New folder/diamonds.csv')
diamond
```

```python
diamond.tail(100)
```

## Scatter plot

```python
# Scatter Plot: Age vs Salary
diamond.plot.scatter(y='carat', x='price', title='Scatter Plot: Carat vs Price (53940 rows Ã— 11 columns) ', 
                              xlabel='Price', ylabel='Carat', grid= True)
```

# Line Plot

```python
diamond.head(500).plot.line(y='carat', x='depth', title='Scatter Plot: carat vs price (500 rows Ã— 11 columns) ', 
                              xlabel='Carat', ylabel='Depth', color='r', marker='o', linestyle='-',grid= True)
```

## Histogram

```python
diamond['carat'].plot.hist(title='carat distribution, 53940 rows Ã— 11 columns',color="yellow")
```

## Boxplot

```python
diamond.boxplot(column='price', by='color', grid=True, vert=False)  # no title
```

## Bar Plot

```python
diamond['clarity'].value_counts().plot.bar(title='Diamond count by clarity ')
```

## Area Plot

```python
diamond.plot.area(y='price', x='color', title='',
                       xlabel='color', ylabel='price', color='g', legend=True, grid=True)
```

## Pie chart

```python
# Pie Chart: Department-wise Salary distribution
diamond.groupby('color')['price'].sum().plot.pie(title='color-wise price Distribution', autopct='%1.1f%%')
```
