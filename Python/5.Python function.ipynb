# Built-in function
A Python function is a block of organized, reusable code that is used to perform a single, related action. Functions provide better modularity for your application and a high degree of code reusing.

```python
#The contents inside the body of the function must be equally indented.##code re-usability
def first_function():
    "this is a docstring"
    print("Hey guys")
    return
```

```python
first_function()
```

```python
def circle(radius):
    print(3.14 * radius* radius)

circle(1.2)
```

```python
def something(text):
    "THIS PRINTS PASSED STRINGS INTO THIS FUNCTION"
    print(text)
```

```python
something("good bad lad chad")
something("lool bool cool fool")
```

```python
def rectangle(a,b):
    return (a*b)/2
rectangle(5,2)
```

```python
c=7
def volume(a,b):
    return (a*b*c)/2
volume(5,2)
```

```python
a=8; b=9; c=12
def volume():
    return (a*b*c)/2
volume()
```
